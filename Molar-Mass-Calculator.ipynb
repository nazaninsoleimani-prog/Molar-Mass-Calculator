## Cell 2 — Test molar mass (no typing needed)

This cell calculates several known compounds. You should see numbers similar to the comments.

````python
examples = [
    ("H2O", 18.015),       # water
    ("H2SO4", 98.072),    # sulfuric acid
    ("NaCl", 58.44),      # table salt
    ("Ca(OH)2", 74.092),  # calcium hydroxide (parentheses!)
    ("Al2(SO4)3", 342.132),
    ("CH3COOH", 60.052),  # acetic acid
]

print(f"{'Formula':<12}{'Calculated':>12}{'Expected':>12}{'OK?':>8}")
print("-" * 44)
for formula, expected in examples:
    got = molar_mass(formula)
    ok = abs(got - expected) < 0.02
    print(f"{formula:<12}{got:>12.3f}{expected:>12.3f}{'YES' if ok else 'NO':>8}")
````

## Cell 3 — Breakdown table

Shows count, atomic mass, contribution, and mass % for each element.

````python
print_breakdown("H2SO4")
print()
print_breakdown("Ca(OH)2")
````

## Cell 4 — See how parsing works

````python
print("H2SO4   ->", parse_formula("H2SO4"))
print("Ca(OH)2 ->", parse_formula("Ca(OH)2"))
print("CH3COOH ->", parse_formula("CH3COOH"))  # C appears twice; counts combine
print("Mass % of H2O:", mass_percent("H2O"))
````

## Cell 5 — Error handling

Bad formulas should print a clear error (not a crash).

````python
bad_inputs = ["", "H2O!", "Xx2", "(OH"]

for bad in bad_inputs:
    try:
        molar_mass(bad)
        print(f"input={bad!r:8} -> unexpected success")
    except ValueError as e:
        print(f"input={bad!r:8} -> Error: {e}")
````
